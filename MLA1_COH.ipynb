```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/weather_2012.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.sample(5)
```

```python
print(df.columns)
```

```python
df =df.rename(columns={'Date/Time':'date_time','Temp (C)':'temp_c','Dew Point Temp (C)':'dew_point','Rel Hum (%)':'relative_humidity','Wind Spd (km/h)':'wind_speed','Visibility (km)': 'visibility', 'Stn Press (kPa)':'pressure','Weather':'weather'})
df
```

```python
df_nums=df.select_dtypes('number')
df_nums.mean()
```

```python
df['temp_f']=df['temp_c']*9/5+32
```

```python
col=df.pop('temp_f')
df.insert(1, 'temp_f',col )
df
```

```python
df.dtypes
```

```python
df['date_time']=pd.to_datetime(df['date_time'])
df
```

```python
monthly_avg=pd.Series(df.groupby(pd.PeriodIndex(df['date_time'], freq='M'))['temp_f'].mean().tolist())
monthly_avg
```

```python
scatter_plot=df.plot.scatter('temp_f','relative_humidity',c='green',title='Temperature vs. Relative Humidity',figsize=(20,10),)
scatter_plot.set_xlabel('Temperature (F)')
scatter_plot.set_ylabel('Relative Humidity')
```

```python
line_plot=df.groupby(pd.PeriodIndex(df['date_time'],freq='m'))['temp_f'].mean().plot.line('date_time','temp_f',c='red',title='Average Temperature per Month')
line_plot.set_xlabel('Month')
line_plot.set_ylabel('Temperature (F)')
```
